OPERADORES LOGICOS /and/or/not

Sirve para cuando queres trabajar con multiples condiciones.
* AND , es verdadero cuando todas las condiciones son verdaderas
* OR , es cuando ocurre una o la otra condicion
* NOT , es cuando invierte la condicion

````python
resultado= (5>3) and (8>6)
print(resultado)
# es verdadero porque ambos son verdaderos

resultado1= (5>3) or (8<6)
print(resultado1)
# Verdadero, porque al menos uno es positivo

resultado2= not (5>3)
print(resultado2)
# es falso porque invierte el resultado
````
